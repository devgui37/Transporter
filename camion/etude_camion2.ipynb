```python
import lib_format_camion as lf
import lib_transport_camion as lt
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = lt.import_data("transport.csv")
```

```python
cost = []
for i in range(1, 11):
    cost.append(lf.total_cout(data, nb_camion=1,capacite=i))
    
cost2 = []
for i in range(1, 11):
    cost2.append(lf.total_cout(data, nb_camion=2,capacite=i))
    
cost3 = []
for i in range(1, 11):
    cost3.append(lf.total_cout(data, nb_camion=3,capacite=i))
```

```python
x = range(1, 11)
y = cost
y2 = cost2
y3 = cost3

fig, ax = plt.subplots()

ax.plot(x, y, linewidth=2.0, color = 'blue', label = '1 camion par entrepôt',alpha = 0.5)
ax.plot(x, y2, linewidth=2.0, color = 'red', label = '2 camions par entrepôt',alpha = 0.5)
ax.plot(x, y3, linewidth=2.0, color = 'green', label = '3 camions par entrepôt',alpha = 0.5)
plt.legend()
plt.xlabel('Capacité des camions (en T)')
plt.ylabel('Coût')
plt.title('Evolution du coût de distribution en fonction de la capacité des camions')
plt.style.use('seaborn')
plt.grid(True)
plt.show()
```
